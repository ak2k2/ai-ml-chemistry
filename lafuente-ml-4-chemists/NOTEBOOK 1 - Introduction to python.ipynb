 As we have pointed out before, float values can be rational or irrational decimal numbers with finite or infinite decimal digits. Computers have a finite computing capacity, which limits the number of decimals the computer stores internally, which means, any mathematical operation we perform will be subject to some sort of approximation. Now, most of the time we can make that approximation as close to the real value as we want and, under certain conditions,  we can we can get exact results. On the other hand, it's convenient to know when a computer is using float values or not, and when it's using approximations or not. For instance, if we want to express the value 0.3, the value that the computer stores and computes might not be exactly 0.3, due to the fact that Python as well as many other programming languages store float values as binary fractions, and most decimal numbers cannot be expressed in binary fractions in an exact way, so they will always be approximations. 
 Nevertheless, when you print a certain float value, Python only prints a *decimal approximation* to the true decimal value of the binary approximation stored by the machine, so you will not notice the difference, however, you might notice the difference when *operating* with these values. For example, when we enter 0.1, the computer operates with the aproximate binary fraction, not with exactly 1/10, so when we add 0.1 + 0.1 + 0.1 the result will not be exactly 0.3.


```python
 .1 + .1 + .1 == .3
```

To start writing a function we start by typing the word "*def*", followed by indentation (whitespaces or tabulations), so it will be easy to recognize.

To understand functions better, we must introduce the concept of **parameter** and **argument**. A *parameter* is a variable that is defined when we declare the function.The parameters are written between parentheses. The *argument* of a given parameter is the actual value of the parameter that gets passed on to the function. 

Aditionally we need to specify what are the **outputs** of our function so we can call them individually as we need them. These outputs are defined as new variables that depend on our previously defined parameters. They are determined by the word "_return_".
Let's re-write our previous math code into a *function* to make this clear:

```python
def function_add (arg1,arg2): 
    # The scope of the variable result is within the function_add and is called local.
    result = arg1 + arg2
    return result

function_add(4,2)
```

```python
def function_addmult (arg1,arg2):
    # The scope of the variable result is within the function_add and is called local.
    result1 = arg1 + arg2
    result2 = arg1*arg2
    return result1, result2

function_addmult(4,2)
```

Let's define a function that allows us to show several tables side-by-side. We'll be doing this for demonstration purposes only, so that we know how to write a function and use it. For this function that we want to create, we will import display_html from IPython.display

```python
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str += df.to_html() # The "+=" operator adds another value with the variable's value and assigns the new value to the variable
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

Inside this function we are defining a new variable: `html_str `, and asigning a blank text value to it, then we're telling the computer to grab the DataFrame (`df`) and render it as an HTML table (same table in a differnt format), adding the table (now in HTML format) as a value to the `html_str` variable, and displaying this side by side with the parameter `'table style="display:inline"'`.
Now we have to "feed" our function with the arguments, which are the tables we want to display:

```python
display_side_by_side(df_I, df_II, df_III,df_IV) 
```

## Linear Regression

A common step on any data analysis is to perform a linear regression. Let's use some native seaborn tools for this purpose.
Let's start with the data on region I. 
The **lmplot** command performs a regression, if we don't specify *which kind* of regression we want, it will output a linear regression.

 **ci=None** is a parameter that gives the order to **not** show the confidence interval (Seaborn shows confidence hyperboles by default).
 We'll show both cases: with and without confidence hyperboles:
 

```python
sns.lmplot(x="x", y="y", data=df[df['dataset']=='I'], ci=None, scatter_kws={"s": 80})
#(the scatter_kws arguments have to do with the size of the points in the graph)
```

```python
sns.lmplot(x="x", y="y", data=df[df['dataset']=='I'], scatter_kws={"s": 80})
```

Let's do the same for region II:

```python
sns.lmplot(x="x", y="y", data=df[df['dataset']=='II'], ci=None, scatter_kws={"s": 80});
```

We see that it's convenvient to do another type of regression, in this case, a parabola will do. We'll have to tell Seaborn to make a second-order polynomial regression. We can accomplish this by using the "order" parameter.

```python
c=sns.lmplot(x="x", y="y", data=df[df['dataset']=='II'], order=2, ci=None,  scatter_kws={"s": 80});
```

### Residuals
It's always useful to make a residuals plot to see if we're using the correct resgression model. Let's keep in mind that if a linear model is adequate for our data, the points in the residual plot should be randomly scattered around the horizontal axis.

```python
sns.residplot(x="x", y="y", data=df[df['dataset']=='I'], scatter_kws={"s": 80});
```

We can do the same for Region II and observe that there's a clear parabolic trend in the points of the residual plot. This indicates that a linear regression is not appropriate for the data.

```python
sns.residplot(x="x", y="y", data=df[df['dataset']=='II'], scatter_kws={"s": 80})
```

###Linear Equation
As we have seen in the previous section, the data in region I does follow a linear trend. Now we'd like to obtain the equation for that trend line. We can do this using the Scipy library by simply plotting region I with the linear regression again, and adding the **plot.legend()** function. If we don't add any arguments to this function, the elements of the legend will be determined automatically.

```python
from scipy import stats

# get coefficients of linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(df[df['dataset']=='I'].x,df[df['dataset']=='I'].y)

# use line_kws to set line label for legend
plot = sns.regplot(x="x", y="y", data=df[df['dataset']=='I'], ci=None, line_kws={'label':"y = {0:.1f} x + {1:.1f}; r = {2:.2f}".format(slope,intercept, r_value)}, scatter_kws={"s": 80})

# plot legend
plot.legend()
```

In order to visualize the linear regression parameters individually with scipy, we can calculate and print them. Here we will also calculate the R squared, p-value and standard error.

```python
# get coefficients of linear fit
slope, intercept, r_value, p_value, std_err, = stats.linregress(df[df['dataset']=='I'].x,df[df['dataset']=='I'].y)
print("slope: %.1f intercept: %.1f Rsquared:  %.2f  p-value: %f standard error: %f " % (slope, intercept,r_value, p_value, std_err))
```
